## 定义一个人脸检测器cv2.CascadeClassifier

```python
face_detector = cv2.CascadeClassifier(r'/Users/zhanghanwen/Opt/anaconda3/lib/python3.9/site-packages/cv2/data/haarcascade_frontalface_alt.xml')
```

```python
## 读取Lena.png图片
img = cv2.imread('Lena.png')
[b, g, r] = cv2.split(img)
img_rgb = cv2.merge([r, g, b])
plt.imshow(img_rgb)
plt.show()
```

## 调用人脸检测器检测人脸位置并展示结果

```python
faces = face_detector.detectMultiScale(img)
```

```python
## 将人脸区域用方框表示出来
for (x,y,w,h) in faces :
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
## 加上文字并展示结果
font=cv2.FONT_HERSHEY_SIMPLEX
img = cv2.putText(img,'Lena',(80,80),font,1.2,(0,0,0),2)
[b, g, r] = cv2.split(img)
img_rgb = cv2.merge([r, g, b])
plt.imshow(img_rgb)
```
