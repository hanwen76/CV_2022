## Pytorch搭建两层全联接神经网络
### 理论推导：
- $\frac{\partial f}{\partial b_2}=-2\left(Y-Y_p\right)^T$
- $\frac{\partial f}{\partial w_2}=-2 h_s^T \cdot\left(Y-Y_p\right)$
- $\frac{\partial f}{\partial b_1}=-2\left(Y-Y_p\right) w_2^T \odot h_s \odot\left(1-h_s\right)$
- $\frac{\partial f}{\partial w_1}=-2 X^T\left(\left(Y-Y_p\right) w_2^T \odot h_s \odot\left(1-h_s\right)\right)$

### 实验过程：

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt


torch.manual_seed(1)    # reproducible

x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)  # x data (tensor), shape=(100, 1)
y = x.pow(2) + 0.2*torch.rand(x.size())       
```

```python
plt.scatter(x.numpy(), y.numpy())
```

### 1. 搭建两层含有bias的全连接网络，隐藏层输出个数为20，激活函数都用sigmoid()

```python
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        #self.W1 = nn.Parameter(nn.init.xavier_normal_(torch.Tensor(n_feature, n_hidden)))
        #self.W2 = nn.Parameter(nn.init.xavier_normal_(torch.Tensor(n_hidden, n_output)))
        #self.b1 = nn.Parameter(nn.init.xavier_normal_(torch.Tensor(100, n_hidden)))
        #self.b2 = nn.Parameter(nn.init.xavier_normal_(torch.Tensor(100, n_output)))
        self.linear1 = torch.nn.Linear(n_feature, n_hidden, bias=True)
        self.linear2 = torch.nn.Linear(n_hidden, n_output, bias=True)

    def forward(self, x):
        outlayer = nn.Sigmoid() # 需要先对类实例化，才能调用
        y_pred = self.linear2(outlayer(self.linear1(x)))
        return y_pred
```

### 2. 实例化神经网络模型

```python
n_feature, n_hidden, n_output = 1, 20, 1
net = Net(n_feature, n_hidden, n_output)     # define the network
print(net)  # net architecture
optimizer = torch.optim.SGD(net.parameters(), lr=0.2)
loss_func = torch.nn.MSELoss()  # this is for regression mean squared loss

plt.ion()   # something about plotting
```

### 3. 训练神经网络，输出结果并作图

```python
for t in range(801):
    prediction = net(x)     # input x and predict based on x
    loss = loss_func(prediction, y)     # must be (1. nn output, 2. target)

    optimizer.zero_grad()   # clear gradients for next train
    loss.backward()         # backpropagation, compute gradients
    optimizer.step()        # apply gradients

    if t % 200 == 0:
        # plot and show learning process
        plt.cla()
        plt.scatter(x.numpy(), y.numpy())
        plt.plot(x.numpy(), prediction.data.numpy(), 'r-', lw=5)
        plt.text(0.5, 0, 't = %d, Loss=%.4f' % (t, loss.data.numpy()), fontdict={'size': 20, 'color':  'red'})
        plt.pause(0.1)
        plt.show()

plt.ioff()
# plt.show()
```
